```python
str2 = list("mississippi")
str1 = list("GAY")
position = 2

final = str1 + str2

# first, shift second half of str1 to the end of final
# then, paste str2 into the open spot
j = len(str1)-1
for i in range(len(str1)+len(str2)-1, position+len(str2)-1, -1):
    final[i] = final[j]
    j -= 1

j = 0
for i in range(position, position+len(str2)):
    final[i] = str2[j]
    j += 1


final = ''.join(final)
print(final)
```
